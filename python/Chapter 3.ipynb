````python
mod = sm.OLS.from_formula('medv ~ lstat', df)
res = mod.fit()
res.summary()
````

````python
new = pd.DataFrame([[1, 5], [1, 10], [1, 15]],
                   columns=['Intercept', 'lstat'])
````

`R`'s `predict(lm)` has a parameter to return the CI for the predicted values. We'll write a helper.

````python
def predict(res, new, interval=True, alpha=.05):
    fit = pd.DataFrame(res.predict(new), columns=['fit'])
    ci = res.conf_int(alpha=alpha).rename(columns={0: 'lower', 1: 'upper'})
    ci = ci.T.dot(new.T).T
    return pd.concat([fit, ci], axis=1)
    
````

````python
predict(res, new)
````
