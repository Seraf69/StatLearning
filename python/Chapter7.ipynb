```python
%matplotlib inline

import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('../data/wage.csv')
df.head()
```

```python
for n in [1, 2, 3, 4]:
    df['age' + str(n)] = df.age**n
```

```python
df.age
mod = sm.OLS.from_formula('wage ~ age + age2 + age3 + age4', df)
res = mod.fit()
res.summary()
```

```python
ax = df.plot(kind='scatter', x='age', y='wage')
ax.scatter(df.age, res.fittedvalues, color='k')
```
